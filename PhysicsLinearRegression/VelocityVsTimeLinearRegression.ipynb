```python
# Imports
import tensorflow as tf
import pandas as pd
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
# Retrieve and Prepare Data
const_accel_data = pd.read_csv('constant_acceleration_measurements.csv')
T = const_accel_data['time']
v = const_accel_data['velocity']
```

```python
# Visualizing Data
velocity_time_graph = plt.plot(T,v)
plt.ylabel('Velocity (m/s)')
plt.xlabel('Time (s)')
plt.show
```

```python
# Simplest Possible Neural Network: 1 Layer that has 1 Nueron and Input Shape is 1 Value
model = keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
```

```python
# Compiling Nueral Network and Specifying 2 Functions
# Loss used to compare guessed answers against known answers and optimizer used to make another guess
# Loss: Mean Squared Error, Optimizer: Stochastic Gradient Descent
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Number of epochs is hyperparameter that indicates the number of passes of the training set the algorithm completed
# Minimizing the loss
model.fit(T, v, epochs=500)
```

```python
# Predicting the velocity at time 14 and 16
predict_cases = [14.00, 16.00]
predictions = model.predict(predict_cases)
```

```python
# Calculating velocity at time 14 and 16 using Kinematics Equation
a = 2.00
v_i = v[0]
calc_times = predict_cases
v_at_time = []
for t in calc_times:
    v_at_time.append(v_i + (a * t))
```

```python
# Comparing prediction to kinematics calculation
for idx in range(len(calc_times)):
    print("Time:", calc_times[idx])
    print("Prediction:", predictions[idx][0]) 
    print("Kinematics Calculation:", v_at_time[idx])
    print()
```
